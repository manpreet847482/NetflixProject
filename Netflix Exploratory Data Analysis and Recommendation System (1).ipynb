# RECOMMENDATION SYSTEM

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')#removing stopwords
```

```python
tfidf_matrix = tfidf.fit_transform(netflix_data['description'])
tfidf_matrix.shape
```

```python
print(tfidf_matrix.shape)
```

```python
from sklearn.metrics.pairwise import linear_kernel  # using sklearn model 
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(netflix_data.index, index=netflix_data['title']).drop_duplicates()
```

```python
def get_recommendations(title):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx])) #Get the pairwsie similarity scores of all movies with that movie
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)#Sort the movies based on the cosine similarity scores
    sim_scores = sim_scores[1:11] #Geting the scores of the 10 most similar movies
    #Started at index 1 because index 0 is the one with the 1 cosine similarity which is the same show
    movie_indices = [i[0] for i in sim_scores]
    return netflix_data['title'].iloc[movie_indices]
```

```python
detail_list= netflix_data['title'].to_list()

def recommend(title, cosine_similarity=cosine_sim): # Defining a funtion to get similiarity for recommended movies and TV shows. 
    index = detail_list.index(title)
    sim_score = list(enumerate(cosine_sim[index]))
    sim_score = sorted(sim_score, key= lambda x: x[1], reverse=True)[1:11] #position 0 is the movie itself, thus exclude
    recommend_index = [i[0] for i in sim_score]
    rec_movie = netflix_data['title'].iloc[recommend_index]
    rec_score = [round(i[1],4) for i in sim_score]
    rec_table = pd.DataFrame(list(zip(rec_movie,rec_score)), columns=['Recommendation','Similarity(0-1)'])
    return rec_table
```
